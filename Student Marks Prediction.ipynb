```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from csv file by using Pandas
df = pd.read_csv("Student Marks Data Set.csv")
df.head(50)
```

```python
# get the data frame infromation by using Pandas info() function
df.info()
```

```python
# get all the statistical information such as count, mean, median, standerd deviation, 25%, 50%,....etc. by using Pandas 
# describe() function
df.describe()
```

```python
cat_columns = df.select_dtypes(include="object").keys()
for i in cat_columns:
    print(df[i].value_counts())
```

```python
# Replace Nan with anyone of the above education type in Parent education column
import random
l2 = []
values = ["Secondary", "Tertiary", "Primary"]  
for i in df["Parent Education"]:
    if pd.isna(i):
        l2.append(random.choice(values))
    else:
        l2.append(i)
df["Parent Education"] = l2
df.head(20)
```

```python
# Convert Categorical data into Numerical data by using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
categorical_columns = df.select_dtypes(include="object").keys()
for i in categorical_columns:
    df[i] = le.fit_transform(df[i])
df.head(20)
```

```python
# Notice the correlation between all columns
df.corr()
```

```python
# Draw the Scatter Plots to notice Linearity
for i in df.keys():
    plt.scatter(df[i], df["Exam_Score"])
    plt.xlabel(i)
    plt.ylabel("Exam_Score")
    plt.show()
```

```python
# Draw the Box-Plots for Observing Outliers
for i in df.keys():
    plt.boxplot(df[i])
    plt.xlabel(i)
    plt.show()
```

```python
# Draw Histograms for Observing Distribution
for i in df.keys():
    plt.hist(df[i])
    plt.xlabel(i)
    plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Split our data into two parts 80% for training 20% for Testing
x_train, x_test, y_train, y_test = train_test_split(df[["HoursStudied/Week"]], df["Exam_Score"], test_size=0.2)

# Create Linear Regression Model
reg = LinearRegression()

# Fit The Model By using Training Data
reg.fit(x_train, y_train)

# get the training score after fit the model 
reg.score(x_train, y_train)
```

```python
# get the Prediction values by using testing score
y_prediction = reg.predict(x_test)

# Calucalting the mean_absolute_error, mean_squared_error, root_mean_squared_error and r2_score
mean_absolute_error = mean_absolute_error(y_test, y_prediction)
mean_squared_error = mean_squared_error(y_test, y_prediction)
root_mean_squared_error = np.sqrt(mean_squared_error)
r2_score = r2_score(y_test, y_prediction)

# Printing all the error values
print("mean_absolute_error", mean_absolute_error)
print("mean_squared_error", mean_squared_error)
print("root_mean_squared_error", root_mean_squared_error)
print("r2_score", r2_score)
```

```python
# Creating data frame for preseant actual values and Predicted values
result_df = pd.DataFrame()
result_df["HoursStudied/Week"] = x_test
result_df["ActualExamScore"] = y_test
result_df["PreictedExamScore"] = y_prediction
result_df.head(10)
```
